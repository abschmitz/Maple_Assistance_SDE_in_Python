```python
import random

N = 1000000
k = 0

for i in range(N):
    Irene = random.random() * 60
    Mike = random.random() * 60
    if abs(Irene - Mike) < 10:
        k += 1

print(k/N)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0.0, 60, 0.01)
y1 = x + 10
y2 = x - 10

fig, (ax) = plt.subplots(1, 1, sharex=True)
ax.set_xlim([0, 60])
ax.set_ylim([0, 60])
plt.xlabel('Irene')
plt.ylabel('Mike')
ax.plot(x, y1, x, y2, color='black')
ax.fill_between(x, y1, y2, where=y2 <= y1, facecolor='red', interpolate=True)
ax.set_title('Irene and Mike Probability')
```
